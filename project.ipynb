<p style="font-size:30px;color:green;font-style: italic;text-decoration:underline;">2. Data Pre-processing</p>

 <p style="font-size:22px;font-style: bold;">Dataset was cleaned by handling missing values, removing duplicates, and correcting column data types. New derived columns like BMI were created, and the data was filtered and aggregated to prepare it for accurate analysis. These steps improved data quality, ensured consistency, and made it easier to identify meaningful patterns related to obesity levels.

<p style="font-size:26px;font-style: italic;">(i) Perform all necessary cleaning 

```python
# 1. Handling missing values
print(df.isnull().sum())   # check missing values
df = df.dropna()           # remove rows with missing values
```

info:- Missing values were checked in the obesity dataset and handled by either filling them with suitable values or removing incomplete rows to ensure accurate analysis.

```python
# 2. Removing duplicates
df = df.drop_duplicates()
```

info:-No duplicate records were found in the obesity dataset, so no rows needed to be removed.

```python
# 3. Correcting data types
df['Age'] = df['Age'].astype(int)     # example
df['Height'] = df['Height'].astype(float)
df['Weight'] = df['Weight'].astype(float)
```

info:-Column data types in the obesity dataset were verified and corrected to proper formats such as numeric and categorical to ensure correct processing and analysis.

```python
# 4. Creating a derived column
df['BMI'] = df['Weight'] / (df['Height'] ** 2)
```

info:- New derived columns, such as BMI, were created from existing fields in the obesity dataset to improve the quality and depth of the analysis.

```python
# 5. Filtering or aggregating data
adults = df[df['Age'] > 18]                               # filter
avg_by_gender = df.groupby('Gender')['Weight'].mean()      # aggregate

print(adults.head())
print(avg_by_gender)
```

info:-Filtering and grouping the obesity dataset helps reveal clearer patterns and trends, making the analysis more meaningful and easier to interpret.

<P STYLE="FONT-SIZE:26PX;font-style: italic;">4. Groupby & Pivot Tables

Computes average BMI by gender.

Creates a pivot table to see average weight across gender and obesity levels.

```python

# Average BMI by Gender
avg_bmi_gender = df.groupby('Gender')['BMI'].mean()
print("Average BMI by Gender:\n", avg_bmi_gender)
```

Analysis:-
The groupby operation calculates the average BMI for each gender. This shows the typical BMI values for males and females in the dataset, allowing comparison between genders.

Conclusion:-
The average BMI is similar for both genders, indicating that gender alone does not cause large differences in obesity levels. This suggests that other factors like diet, physical activity, and genetics may play a stronger role in determining BMI.

```python
# Pivot table: Average Weight by Gender and Obesity Category
pivot_table = df.pivot_table(
    values='Weight', 
    index='Gender', 
    columns='NObeyesdad', 
    aggfunc='mean'
)

print("\nPivot Table (Weight by Gender and Obesity Category):\n", pivot_table)


```

Analysis:-The pivot table shows the average weight of males and females across different obesity categories. It helps compare how weight varies with both gender and obesity class.

Conclusion:-Weight differences are visible across obesity categories, and males and females show slightly different patterns. This confirms that gender and obesity level together influence body weight, which is useful for understanding risk factors.

<p style="font-size:26px;font-style: italic;">5.Correlation Analysis

Calculates correlations between numerical columns.

Shows a heatmap for easy visualization.

```python
correlation_matrix = df[['Age','Height','Weight','BMI']].corr()
print("\nCorrelation Matrix:\n", correlation_matrix)

# Heatmap for correlation
plt.figure(figsize=(6,4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

 Analysis:-
The correlation matrix shows the strength and direction of relationships between numerical variables: Age, Height, Weight, and BMI.
Values close to 1 indicate strong positive correlation, 0 indicates no correlation, and -1 indicates strong negative correlation.
The heatmap visually highlights these relationships. From the dataset, Weight and BMI show a very strong positive correlation, while Height has a weaker positive correlation with BMI. Age shows little correlation with BMI and Weight.

 Conclusion:-
The correlation analysis confirms that BMI is strongly influenced by Weight, and less by Height or Age. This insight is useful for identifying key features when analyzing or predicting obesity levels.
